```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset("titanic")

titanic_age_sex = titanic[['age', 'sex']].dropna(subset=['age']) 

sns.histplot(data=titanic_age_sex, x='age', hue='sex', bins=20, kde=False, multiple='dodge')

plt.title("Age Distribution by Gender - Titanic Passengers")
plt.xlabel("Age")
plt.ylabel("Count")

plt.show()
```
